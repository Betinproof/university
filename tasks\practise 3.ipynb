```python
import random

class Game:
    def __init__(self, words, lives):
        self.words = words
        self.lives = lives
        self.guessed_words = set()

    def choose_word(self):
        return random.choice(self.words)

    def display_word(self, word, guessed_letters):
        displayed_word = ''
        for letter in word:
            if letter in guessed_letters:
                displayed_word += letter
            else:
                displayed_word += "\u25A1"
        return displayed_word

    def play_game(self):
        word_to_guess = self.choose_word()
        guessed_letters = set()

        while self.lives > 0:
            display = self.display_word(word_to_guess, guessed_letters)
            print("Current Word: ", display)

            guess = input("Enter a letter or the whole word: ").lower()

            if guess == word_to_guess:
                print("Congratulations! You guessed the word!")
                self.guessed_words.add(word_to_guess)
                break
            elif guess.isalpha() and len(guess) == 1:
                guessed_letters.add(guess)
                if set(word_to_guess) == guessed_letters:
                    print("Congratulations! You guessed the word!")
                    self.guessed_words.add(word_to_guess)
                    break
            else:
                print("Incorrect guess. Try again.")
                self.lives -= 1

        return word_to_guess in self.guessed_words
```

```python
from game_logic import Game

def choose_difficulty():
    difficulty = input("Choose difficulty (easy, medium, hard): ").lower()
    if difficulty == 'easy':
        return 7
    elif difficulty == 'medium':
        return 5
    elif difficulty == 'hard':
        return 3
    else:
        print("Invalid difficulty. Defaulting to medium.")
        return 5

def play():
    words = ["книга", "месяц", "ручка", "шарик", "олень", "носок"]
    lives = choose_difficulty()

    game = Game(words, lives)
    while words:
        result = game.play_game()
        if result:
            words.remove(result)
        replay = input("Do you want to play again? (yes/no): ").lower()
        if replay != 'yes':
            break

if __name__ == "__main__":
    play()
```
